# Графики выявленных случаев COVID-19 
### для городов/штатов: Москва, Лондон, Мадрид, Амстердам, Калифорния и Иллинойс

```python
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
file_Moscow = open('Moscow_covid_cases.csv', 'r')
Moscow_date_list = []
Moscow_Cases_list = []
for line in file_Moscow:
    date = datetime.strptime(line.split()[0], '%d.%m.%Y')
    Cases = int(line.split()[1])
    Moscow_date_list.append(date)
    Moscow_Cases_list.append(Cases )
file_Moscow.close()
```

```python
file = open('London_covid_cases.csv', 'r')
London_date_list = []
London_Cases_list = []
for line in file:
    date = datetime.strptime(line.split()[0], '%d.%m.%Y')
    Cases = int(line.split()[1])
    London_date_list.append(date)
    London_Cases_list.append(Cases )
file.close()
```

```python
file = open('Illinois_covid_cases.csv', 'r')
Illinois_date_list = []
Illinois_Cases_list = []
for line in file:
    date = datetime.strptime(line.split()[0], '%d.%m.%Y')
    Cases = int(line.split()[1])
    Illinois_date_list.append(date)
    Illinois_Cases_list.append(Cases)
file.close()
```

```python
fig, axes = plt.subplots(figsize=(12,4), dpi=300)
# Add set of axes to figure
plt.grid(True)
plt.plot(Moscow_date_list,Moscow_Cases_list, label = "Moscow")
plt.plot(London_date_list,London_Cases_list, label = "London")
plt.plot(Madrid_date_list,Madrid_Cases_list, label = "Madrid")
plt.plot(Amsterdam_date_list,Amsterdam_Cases_list, label = "Amsterdam")
plt.plot(California_date_list,California_Cases_list, label = "California")
plt.plot(Illinois_date_list,Illinois_Cases_list, label = "Illinois")
plt.xlabel('Дата')
plt.ylabel('Количество заражений')
plt.title('Число новых заражений')
plt.legend()
plt.gcf().autofmt_xdate()

plt.show()
```
